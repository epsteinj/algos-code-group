```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Average(n):
    pi = open('pi', 'r')
    digits = np.array(list(pi.read(n)), dtype=int)
    mean = np.mean(digits)
    return mean
```

```python
Average(1000000000)
```

```python
x = list(range(1, 1001))
y = list(map(Average, x))
```

```python
plt.autoscale()
plt.scatter(x, y, s=5)
plt.title('Average of Pi vs Number of Digits of Pi')
plt.xlabel('# of digits of pi')
plt.ylabel('average')
```
